```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(0, 10, 100)#guna numpy
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots() #buatcanvas
ax.plot(x, y1, color='blue', label='Sin')
ax.plot(x, y2,  color='red', label='Cos')

ax.set_title("Perbandingan Sin & Cos")
ax.legend()
plt.show()
```

```python
# Fungsi untuk menentukan status lulus/gagal
def semak_status(markah):
    if markah >= 50:
        return "LULUS"
    else:
        return "GAGAL"

# Dictionary untuk menyimpan data kelas
data_kelas = {}

print("=== SISTEM PENGURUSAN MARKAH KELAS ===")
try:
    bil_pelajar = int(input("Masukkan bilangan pelajar dalam kelas: "))
    print("-" * 40)

    # Gelung untuk mengambil data setiap pelajar
    for i in range(bil_pelajar):
        nama = input(f"Masukkan nama pelajar ke-{i+1}: ").strip().capitalize()

        while True:
            try:
                markah = float(input(f"Masukkan markah untuk {nama} (0-100): "))
                if 0 <= markah <= 100:
                    break
                print("Ralat: Markah mestilah antara 0 hingga 100.")
            except ValueError:
                print("Ralat: Sila masukkan nombor yang sah.")

        # Simpan dalam dictionary
        data_kelas[nama] = markah
        print("-" * 40)

    # Proses Analisis Data (Jika ada data)
    if data_kelas:
        jumlah_markah_kelas = sum(data_kelas.values())
        purata_kelas = jumlah_markah_kelas / len(data_kelas)

        # Mencari markah tertinggi dan terendah
        pelajar_terbaik = max(data_kelas, key=data_kelas.get)
        pelajar_terendah = min(data_kelas, key=data_kelas.get)

        # Paparan Rumusan Keseluruhan
        print("\n" + "="*50)
        print("                LAPORAN MARKAH KELAS")
        print("="*50)
        print(f"{'NAMA PELAJAR':<20} | {'MARKAH':<10} | {'STATUS':<10}")
        print("-" * 50)

        for nama, markah in data_kelas.items():
            status = semak_status(markah)
            print(f"{nama:<20} | {markah:<10.2f} | {status:<10}")

        print("-" * 50)
        print(f"Purata Markah Kelas : {purata_kelas:.2f}")
        print(f"Pelajar Terbaik     : {pelajar_terbaik} ({data_kelas[pelajar_terbaik]:.2f})")
        print(f"Pelajar Terendah    : {pelajar_terendah} ({data_kelas[pelajar_terendah]:.2f})")
        print("="*50)

except ValueError:
    print("Ralat: Sila masukkan bilangan pelajar dalam bentuk nombor bulat.")
```
